```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
data = np.zeros((120,3))
x1 = np.random.normal(0,0.6,100)
y1 = np.random.normal(0,0.6,100)
x2 = np.random.normal(2,0.8,20)
y2 = np.random.normal(2,0.8,20) 

for i in range(100):
    data[i,0] = x1[i]
    data[i,1] = y1[i]
    data[i,2] = 1
    
for i in range(20):
    data[i+100,0] = x2[i]
    data[i+100,1] = y2[i]
    data[i+100,2] = -1
```

```python
w1,w2,b,alpha = np.random.normal(0,0.1,4)
alpha = math.fabs(alpha)
temp = np.ones((120,1))
epoch=0
error = []
```

```python
while(epoch < 1000):
    
    for i in range(120):
        net = w1*data[i,0]+w2*data[i,1]+b
        w1 += alpha*(data[i,2]-net)*data[i,0]
        w2 += alpha*(data[i,2]-net)*data[i,1]
        b += alpha*(data[i,2]-net)
        
    for i in range(120):
        temp[i] = math.fabs(w1*data[i,0]+w2*data[i,1]+b-data[i,2])**2*0.5
        
    error.append(temp.sum()/120)
        
    epoch += 1
```

```python
x = np.linspace(-2,2,3)
y = -(w1*x+b)/w2
plt.plot(x,y,'k')
x1 = data[:100,0]
y1 = data[:100,1]
x2 = data[100:,0]
y2 = data[100:,1]
plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.grid()
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.figure()
plt.plot(error)
plt.xlabel('number of steps')
plt.ylabel('0.5*(t-net)^2')
```
